```python
import pandas as pd
```

## reading dataset

```python
df=pd.read_csv("worldPopulation.csv", skiprows=4)
```

```python
df.head()
```

```python
print(df.head())
```

## importing necessary librariy

```python
import matplotlib.pyplot as plt
```

# Population Bar chart for first 10 countries 

```python
plt.figure(figsize=(15,8))
df1=df.head(10)
plt.bar(df1['Country Name'],df1['2023'])
plt.title('Population Distribution by country (2023)',fontsize=10)
plt.xlabel('Country',fontsize=10)
plt.ylabel('Population',fontsize=10)
plt.xticks(rotation=90)
plt.show()
```

# Population distribution using seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,50))
sns.barplot(x='2023',y='Country Name',data=df)
plt.title('Population distribution by country(2023)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

# Bar chart of selected countries

```python
selected_countries = ['India', 'China', 'United States']  
# Filter the DataFrame to include only the selected countries
filtered_df = df[df['Country Name'].isin(selected_countries)]
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2,2,1)
plt.bar(filtered_df['Country Name'], filtered_df['2023'])
plt.title('Population Distribution by Selected Countries (2023)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45) 
plt.show()
```

# Histogram of Population Distribution in 2023

```python
# Drop rows with missing population values
df = df.dropna(subset=['2023'])

# Plotting the histogram
plt.figure(figsize=(12, 10))
n, bins, patches = plt.hist(df['2023'], bins=10, edgecolor='black')

# Adding the population bin range as x-axis labels
plt.xticks(ticks=bins, labels=[f'{int(b)}' for b in bins], rotation=90)
plt.title('Histogram of Population Distribution (2023)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```
